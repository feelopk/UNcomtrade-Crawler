## Network Top rank change?

### Import (Before 2020)

```python
Quantile = 0.2
before_imG = nx.read_edgelist('%s/Im_Top%s_before2020_mean_kg.csv' %(data_read_path, Quantile),\
                      create_using = nx.DiGraph(), delimiter = '\t', \
                    comments = 'Source', data = (("weight", float),))

before_im_cendf = pd.DataFrame(index = before_imG.nodes()) 
print(len(list(before_imG.nodes())))
ind_cen = nx.in_degree_centrality(before_imG) # no weight
eig_cen = nx.eigenvector_centrality(before_imG, weight = 'weight') # weight 
katz_cen = nx.katz_centrality(before_imG, weight = 'weight') # weight 

''' add to df '''
before_im_cendf['Indgr_cen'] = before_im_cendf.index.map(ind_cen)
before_im_cendf['Eig_cen'] = before_im_cendf.index.map(eig_cen)
before_im_cendf['Katz_cen'] = before_im_cendf.index.map(katz_cen)
before_im_cendf.to_csv('%s/Im_Before2020_Top%s_centrality.csv' %(data_read_path, Quantile), sep = '\t')
```
